# 1. Three (3) Components of Recursion

- Component 1: **Terminating Conditioning** (aka. **Base Case**) that returns without making a recursive call.
    - e.g. an if condition statement to return when trigger a condition (e.g. count tracker in component 2 <=0).
    - without it, there will be infinite recursion -> which will trigger "**StackOverFlow**" error, which is a subclass of "Error" class
    
    

- Component 2: **Reduction Step** that moves the direction of the recursive calls towards the terminating conditioning above.
    - e.g. a count tracker where every recursion is called, the count - 1.
- Component 3: **Recursive Call**
    - e.g. call the method within the method itself.
    - Recursion takes advantage of Java's "**Stack**"-based nature.
