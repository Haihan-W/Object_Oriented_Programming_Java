## For Statement


VIP NOTE1!!!!!!
- The for statement evaluate the body first, then update the counter, then enter the next loop, same as while statement:
    - So in the example 1 below, the lineNum when printing the first "Hello, world" is 1, instead of 2;
    - And (still example 1 below) the lineNum when printing the 10th "Hello, World" is 10. Then after printing, lineNum is updated to 11 by lineNum++, then iterate back to for loop to check condition, and find lineNum<=10 is not satisfied, then exit the loop.
        - So the lineNum++ before exiting the loop is 11 instead of 10
