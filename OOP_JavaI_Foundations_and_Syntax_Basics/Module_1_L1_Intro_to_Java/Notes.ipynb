## Basic Java Program




Example:

- HelloWorld{} is a class
- main() is a method
    - This method will be automatically called when running the program
- println() is a method
    - Different from main(), Most methods in Java like this needs to be explicitly called in order to execute

## Compiler Language vs. Interpreter Language

Java -> Hybrid Approach


- Java compiled high level language to bytecode (instead of machine code as C,C++) and stored in .class extension (low-level language)
- Then using interpreter to translage bytecode to machine language of particular processor of the target computer.
- The java interpreter is often referred to as **"Java Virtual Machine" (JVM)**
    - The “Virtual Machine” part of the name is derived from the fact that compiled Java code is not executed by a real processor. Rather a piece of software, the interpreter, performs the execution.

## Java Development Kit (JDK)

To access the necessary tools for translating and executing Java programs using the terminal, install the Java Development Kit (or JDK).  The JDK is a large software product that consists of several tools for creating, executing, and managing Java programs.  Appropriately, the compiler and interpreter are two such (very important) components.

## Compile and Execute Java Program

Step 1. Compile. At the end, there will be a .class file generated
    
    $ javac HelloWorld.java

Step 2. Execute. (Please note: do not include file extension!!!)
    
    $ java HelloWorld
