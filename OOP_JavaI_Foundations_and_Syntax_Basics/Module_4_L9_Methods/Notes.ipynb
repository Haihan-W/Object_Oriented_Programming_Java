# User Defined Methods

## 1. Format:
- Three Elements are required for **Methods Header**:
    - Return Type
        - which represents a type of data the method will return to its caller.
        - Not all methods need a return value, if not need a return, then "void" is used as Return Type
    - Method Name
    - Formal Parameters
        - which represents input(s) that you can actually use within the method
        - Not all methods need an input, if not need input, simply have a set of parentheses with no parameters in the header, e.g. methodtype methodname()
        
        
- Optional Elements for **Methods Header**:
    - Modifiers:
        - Ex1: "Public" modifier
            - Public is known as a visibility modifier.
            - Including it here will allow us to call the method searchStringArray in another **class** outside of this current one that we're writing.
        - Ex2: "Static" modifier
            - We'll explore the concept of static versus non-static methods much more deeply in the next couple of sections.
            - For now we'll just use "Static" for each method we define inside a class that won't have objects of its type, which means In other words, we won't be instantiating ArraySearch objects, rather the class will be a container of an ArraySearch method.

## 3. Method Overload

- To overload a method, simply:
    - a. same method name
    - b. different formal parameters for input (i.e. different types, and/or order, and/or number of formal parameters) 
    - Note: doesn't matter if the return type or modifier is the same or not; Also return type can both be void if no return

    - Below each pair of methods does NOT have method overload, because the only difference is return type, there is not difference for input (formal parameters)
        - Methods within Each pair CANNOT co-exist because since their input is the same, there is no way to tell which method to use when method is invoked.
        - The Java program will simply throw an error in compile stage, saying method is already defined and cannot be defined twice.
    
    
    
    
- Why to overload method:
    - In this case, we can avoid using multiple different method names if only the difference is the input (e.g. input type) and the method is similar to each other.
